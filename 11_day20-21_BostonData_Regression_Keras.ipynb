```python
from keras.datasets import boston_housing

(train_data,train_target),(test_data,test_labels) = boston_housing.load_data()

train_data.shape
test_data.shape
```

```python
mean = train_data.mean(axis=0)
train_data -= mean
std = train_data.std(axis=0)
train_data /= std
test_data -= mean
test_data /= std
 
```

```python
train_labels
```

```python
from  keras import models
from  keras import layers,optimizers

def build_model():
    model = models.Sequential()
    model.add(layers.Dense(64,activation='relu',input_shape=(train_data.shape[1],)))
    model.add(layers.Dense(64,activation='relu'))
    model.add(layers.Dense(1))

    model.compile(optimizer='rmsprop',loss='mse',metrics=['mae'])
    return model

```

```python
import numpy as np
k=4

num_val_sample = len(train_data)//4
num_val_sample
```

```python
num_epochs=20
all_mse_scores =[]
all_mae_scores = []

for i in range(k):
    print('----Fold  :',i)
    val_data = train_data[i*num_val_sample:(i+1)*num_val_sample]
    val_target = train_target[i*num_val_sample:(i+1)*num_val_sample]
    
    org_train_data = np.concatenate((train_data[(i+1)*num_val_sample:],train_data[:i*num_val_sample]))
    org_train_target = np.concatenate((train_target[(i+1)*num_val_sample:],train_target[:i*num_val_sample]))
    print('val_data ',val_data.shape)
    
    print('org_train_data ',org_train_data.shape)
     
    
    model = build_model()
    model.fit(org_train_data,org_train_target,epochs=num_epochs,batch_size=1)
    val_mse,val_mae = model.evaluate(val_data,val_target)
    all_mae_scores.append(val_mae)
    
print('All mae scores',all_mae_scores)
print('Mean of all mae scores',np.mean(all_mae_scores))
    
    
```
